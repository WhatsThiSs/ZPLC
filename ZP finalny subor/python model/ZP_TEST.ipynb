```python
import pandas as pd

# Načítanie údajov zo súborov
dotaznik1_updated = pd.read_csv('dotaznik1_updated.csv')
dotaznik2_updated = pd.read_csv('dotaznik2_updated.csv')
level1_updated = pd.read_csv('level1_updated.csv')
level2_updated = pd.read_csv('level2_updated.csv')
level3_updated = pd.read_csv('level3_updated.csv')

# Spojenie všetkých načítaných dátových rámcov do jedného
data = pd.concat([dotaznik1_updated, dotaznik2_updated, level1_updated, level2_updated, level3_updated], ignore_index=True)

# Zistenie počtu jedinečných záznamov a počtu duplikátov
pocet_jedinecnych = data.drop_duplicates().shape[0]
pocet_duplikatov = data.shape[0] - pocet_jedinecnych

# Spočítanie počtu chýbajúcich hodnôt
pocet_chybajucich = data.isnull().sum().sum()

# Výpis výsledkov
print(f"Počet jedinečných záznamov: {pocet_jedinecnych}")
print(f"Počet duplikátov: {pocet_duplikatov}")
print(f"Počet chýbajúcich hodnôt: {pocet_chybajucich}")
```
