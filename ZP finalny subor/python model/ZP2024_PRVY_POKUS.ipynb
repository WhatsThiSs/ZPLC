```python
def map_to_emotion(value, field):
    if field == 'expectation_words':
        if value == 6:
            return 'Nadšenie'
        elif value == 5:
            return 'Radosť'
        elif value == 4:
            return 'Spokojnosť'
        elif value == 3:
            return 'Neutralita'
        elif value == 2:
            return 'Úzkosť'
        elif value == 1:
            return 'Hnev'
        elif value == 0:
            return 'Strach'
    elif field == 'current_emotion':
        if value >= 7:
            return 'Nadšenie'
        elif value == 5 or value == 6:
            return 'Radosť'
        elif value == 4:
            return 'Spokojnosť'
        elif value == 3:
            return 'Neutralita'
        elif value == 2:
            return 'Úzkosť'
        elif value == 1:
            return 'Hnev'
        elif value == 0:
            return 'Strach'
    else:
        return 'Neznáma'
```

```python
import pandas as pd

dotaznik1_data = pd.read_csv("dotaznik1_test.csv")
dotaznik2_data = pd.read_csv("dotaznik2_test.csv")

# Aplikovanie mapovacej funkcie na stĺpce 'expectation_words' a 'current_emotion' v oboch dátových rámcov
dotaznik1_data['expectation_words_emotion'] = dotaznik1_data['expectation_words'].apply(lambda x: map_to_emotion(x, 'expectation_words'))
dotaznik1_data['current_emotion_emotion'] = dotaznik1_data['current_emotion'].apply(lambda x: map_to_emotion(x, 'current_emotion'))

dotaznik2_data['expectation_words_emotion'] = dotaznik2_data['expectationsafter'].apply(lambda x: map_to_emotion(x, 'expectation_words'))
dotaznik2_data['current_emotion_emotion'] = dotaznik2_data['current_emotion'].apply(lambda x: map_to_emotion(x, 'current_emotion'))

# Zobrazenie emocionálnych stavov pred a po
print("Dotazník 1:")
print(dotaznik1_data[['expectation_words_emotion', 'current_emotion_emotion']])
print("\nDotazník 2:")
print(dotaznik2_data[['expectation_words_emotion', 'current_emotion_emotion']])
```

```python
merged_data = pd.concat([dotaznik1_data, dotaznik2_data], keys=['Dotaznik1', 'Dotaznik2'])
# Zistenie počtu zlepšení a zhoršení pre každú emóciu
improvement_count = merged_data.apply(lambda x: x['current_emotion_emotion'] if x['current_emotion_emotion'] == x['expectation_words_emotion'] else None, axis=1).groupby(level=0).count()
deterioration_count = merged_data.apply(lambda x: x['current_emotion_emotion'] if x['current_emotion_emotion'] != x['expectation_words_emotion'] else None, axis=1).groupby(level=0).count()
# Výpočet percentuálnych zmien
total_count = merged_data.groupby(level=0).size()
improvement_percentage = (improvement_count / total_count) * 100
deterioration_percentage = (deterioration_count / total_count) * 100
# Zobrazenie výsledkov
change_summary = pd.DataFrame({'Zlepšenie (%)': improvement_percentage, 'Zhoršenie (%)': deterioration_percentage})
print(change_summary)
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Preddefinované hodnoty váh pre jednotlivé vstupné premenné
weights = {
    'mouseexp': 0.2,
    'handedness': 0.1,
    'gender': 0.1,
    'name': 0.1,
    'age': 0.1,
    'important': 0.1,
    'factor': 0.3,
    'expectation_words_x': 0.25,
    'current_emotion_x': 0.35,
    'TimeRemaining_x': 0.2,
    'Misses_x': 0.15,
    'Hits_x': 0.25,
    'overallAverageMouseSpeed_x': 0.3,
    'thresholdedDirectionChanges_x': 0.4,
    'ClickFrequency_x': 0.35,
    'Level_x': 0.25,
    'TimeRemaining_y': 0.2,
    'Misses_y': 0.15,
    'Hits_y': 0.25,
    'overallAverageMouseSpeed_y': 0.3,
    'thresholdedDirectionChanges_y': 0.4,
    'ClickFrequency_y': 0.35,
    'Level_y': 0.25,
    'expectationsafter': 0.3,
    'hard': 0.25,
    'changes': 0.2,
    'expectation_words_y': 0.25,
    'expectationwordafter': 0.35,
    'current_emotion_y': 0.4,
    'Misses': 0.15,
    'Hits': 0.25,
    'overallAverageMouseSpeed': 0.3,
    'thresholdedDirectionChanges': 0.4,
    'ClickFrequency': 0.35,
    'Level': 0.25
}

# Nahranie dát z premennej final_merged_data
X_train = final_merged_data.drop(columns=['current_emotion_x', 'Session', 'name']).values  # Vyber všetkých stĺpcov okrem current_emotion_x, Session a name
y_train = final_merged_data['current_emotion_x'].values  # Definovanie cieľovej premennej

# Pridanie chýbajúcej váhy pre kľúč 'Session'
if 'Session' not in weights:
    weights['Session'] = 1.0


# Normalizácia údajov s použitím váh
X_train_scaled = X_train.copy()  # Skopírovanie údajov na normalizáciu

# Normalizácia údajov s použitím váh pre každý stĺpec
for col in range(X_train.shape[1]):
    col_name = final_merged_data.columns[col]
    weight = weights.get(col_name, 1.0)  # Získanie váhy z dictionary
    if isinstance(weight, str):  # Skontrolujeme, či je váha reťazec
        weights[col_name] = float(weight)  # Ak áno, skonvertujeme ju na float
    X_train_scaled[:, col] *= weights[col_name]  # Násobenie váhy

# Normalizácia údajov
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_scaled)

# Rozdelenie údajov na trénovaciu a testovaciu množinu
X_train, X_test, y_train, y_test = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=42)

# Definícia modelu neurónovej siete
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Kompilácia modelu
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Trénovanie modelu
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)

# Evaluácia modelu na testovacej množine
loss, mae = model.evaluate(X_test, y_test)
print(f'Test loss: {loss}, Test MAE: {mae}')
```
