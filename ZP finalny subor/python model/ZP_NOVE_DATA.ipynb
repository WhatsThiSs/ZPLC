```python
# RANDOMFOREST 
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Načítanie dát
data = pd.read_csv('data_merged_AA.csv')  

# Pridanie ďalších atribútov do dát
additional_features = [ 
      
        'current_emotion_y',
       'overallAverageMouseSpeed_x',
       'thresholdedDirectionChanges_x',  'TimeRemaining_y',
       'overallAverageMouseSpeed_y',
       'thresholdedDirectionChanges_y', 'ClickFrequency_y', 
       'Misses', 'Hits', 'overallAverageMouseSpeed',
       'thresholdedDirectionChanges', 'ClickFrequency']   

X = data[additional_features]
y = data['expectationwordafter']  # Cieľová premenná

# Normalizácia dát
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Rozdelenie dát na trénovaciu a testovaciu sadu
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Vytvorenie a trénovanie modelu Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predikcia na testovacej sade
y_pred = rf_model.predict(X_test)

# Vyhodnotenie modelu pomocou presnosti (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print("Presnosť modelu:", accuracy)

# Zistenie dôležitosti príznakov
feature_importance = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': additional_features, 'Importance': feature_importance})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score

# Načítanie dát
data = pd.read_csv('updated_data_merged.csv')  

# Pridanie ďalších atribútov do dát
additional_features = [ 
    'current_emotion_y',
    'overallAverageMouseSpeed_x',
    'thresholdedDirectionChanges_x',  
    'TimeRemaining_y',
    'overallAverageMouseSpeed_y',
    'thresholdedDirectionChanges_y', 
    'ClickFrequency_y', 
    'Misses', 'Hits', 
    'overallAverageMouseSpeed',
    'thresholdedDirectionChanges', 
    'ClickFrequency'
]   

X = data[additional_features]
y = data['new_expectation']  # Cieľová premenná

# Normalizácia dát
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Rozdelenie dát na trénovaciu a testovaciu sadu
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Vytvorenie a trénovanie modelu Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predikcia na testovacej sade
y_pred = rf_model.predict(X_test)

# Vyhodnotenie modelu pomocou presnosti (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print("Presnosť modelu:", accuracy)

# Vyhodnotenie modelu pomocou F1 skóre
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1 skóre modelu:", f1)

# Zistenie dôležitosti príznakov
feature_importance = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': additional_features, 'Importance': feature_importance})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)
```
